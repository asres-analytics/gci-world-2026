## 📘 Introduction to NumPy


NumPy (Numerical Python) is a core Python library designed for **efficient numerical and scientific computing**.

## Vectorization

`Vectorization` is the practice of applying operations to entire NumPy arrays at once,
instead of processing elements one by one with Python loops. This results in cleaner,
shorter, and more efficient code.




One of the most important concepts I learned was **vectorization**.

Instead of using explicit Python loops, NumPy applies operations **element‑by‑element** using optimized C‑level code (universal functions).
This results in:
- Cleaner and more readable code
- Significant performance improvements
- Fewer logical errors


### Example: Vectorized Operation

````python

a = np.array([1, 2, 3, 4, 5])

# Vectorized operation (no loop)
b = a * 2
b
````


Vectorization applies operations element‑by‑element at C‑level speed, resulting in cleaner code and better performance compared to explicit Python loops.

This replaces loop‑based code such as:

```python for i in range(len(a)):
    a[i] = a[i] * 2
```

and demonstrates how NumPy simplifies numerical transformations.


## Universal functions

`Universal functions` *`(ufuncs)`* are the underlying NumPy mechanisms that make vectorization
possible. They are implemented at the C‑level and apply operations element‑by‑element
across arrays, providing high performance and automatic support for broadcasting.



### Examples: Universal functions


````python
data = np.array([1, 4, 9])
np.add(data, 5) # add 5 to each element in the array and output is [ 6, 9, 14]
np.subtract(data, 2) # subtract 2 from each element in the array and output is [-1, 2, 7]
np.multiply(data, 3)  # multiply each element in the array by 3 and output is [ 3, 12, 27]
np.divide(data, 2)    # divide each element in the array by 2 and output is [ 0.5, 2. , 4.5]
np.sqrt(data)     # calculate the square root of each element in the array and output is [1. 2. 3.]
np.exp(data.round(2))) # calculate the exponential of each element in the array and output is [ 2.71828183, 54.59815003, 8103.08392758]
````

In summary, vectorization describes **how NumPy code is written**, while universal
functions describe **how NumPy executes that code efficiently**.

## Broadcasting

Broadcasting allows NumPy to perform operations on arrays of different shapes automatically, without copying data.

I learned how NumPy:
- Expands scalars to match array shapes
- Applies operations efficiently
- Raises errors when shapes are incompatible

### Example: Scalar Broadcasting

````python

arr = np.array([[1, 2], [3, 4]])
arr + 10 # add 10 to each element in the array and output is [[11, 12], [13, 14]]
arr * 2 # multiply each element in the array by 2 and output is [[ 2, 4], [6, 8]]
arr ** 2 # square each element in the array and output is [[ 1, 4], [9, 16]]    

````


Broadcasting allows NumPy to automatically align array shapes, enabling operations between arrays and scalars without copying data.


## Indexing and Data Selection

Indexing and slicing allow precise selection of elements and sub‑arrays.
NumPy uses **zero‑based indexing**, similar to Python lists.

###  Example: Indexing and Slicing

````python
arr = np.array([10, 20, 30, 40, 50])

arr[2]        # single element at index 2 and output is 30
arr[1:4]      # slicing with start at index 1 and end at index 4 (exclusive) and output is [20, 30, 40]
arr[::2]      # slicing with step of 2 and output is [10, 30, 50]
arr[-1]       # negative indexing to access last element and output is 50   
````


Indexing and slicing allow precise extraction of elements and sub‑arrays. I learned that slicing returns **views**, not copies, which is memory‑efficient.


##  Working with 2D Arrays & Axis

Two‑dimensional NumPy arrays are commonly used to represent **tabular data**, where
***rows*** correspond to **`observations`** and ***columns*** correspond to **`features`**.

Most operations on 2D arrays are applied element‑wise by default. To perform
meaningful summaries across rows or columns, NumPy provides the `axis` parameter,
which controls the direction of computation.

- `axis=0` operates **column‑wise** (down the rows)
- `axis=1` operates **row‑wise** (across the columns)

Understanding how the `axis` parameter works is essential for correctly aggregating
and analyzing structured numerical data.

### Example: Working with 2D Arrays & Axis

````python
# create a 2D array and calculate mean along different axes 

data = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

np.mean(data, axis=0)  # column-wise mean and output is [25. 35. 45.]
np.mean(data, axis=1)  # row-wise mean and output is [20. 50.]
````

 ### axis=0 → column-wise
Computes the mean down each column:

Column 0: (10 + 40) / 2 = 25

Column 1: (20 + 50) / 2 = 35

Column 2: (30 + 60) / 2 = 45

### axis=1 → row-wise
Computes the mean across each row:

Row 0: (10 + 20 + 30) / 3 = 20

Row 1: (40 + 50 + 60) / 3 = 50


Using the `axis` parameter helped me distinguish between **row‑wise** and **column‑wise** aggregation, which is critical for tabular data analysis.


## Reshaping

**Reshaping** changes the **structure (shape)** of a NumPy array without changing
the underlying data. It allows the same values to be viewed in a different
arrangement, as long as the total number of elements remains the same.

Reshaping is commonly used when preparing data for analysis, matrix operations,
or machine learning models that require data in a specific dimensional format.

### Example: Reshaping a 1D Array into 2D

````python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr)
print("Shape:", arr.shape)
````

This means:

- 1D array
- 6 elements

### Reshape into 2D
- Convert into 2 rows, 3 columns

````python
arr_2d = arr.reshape(2, 3)
print(arr_2d)
print("Shape:", arr_2d.shape)
````

- Convert into 3 rows, 2 columns

````python
arr_2d = arr.reshape(2, 3)
print(arr_2d)
print("Shape:", arr_2d.shape)
````

## Important Rule 

 Total elements must match
  - 6 elements = 2 × 3 ✔

  - 6 elements = 3 × 2 ✔

  - 6 elements ≠ 4 × 2 ❌
