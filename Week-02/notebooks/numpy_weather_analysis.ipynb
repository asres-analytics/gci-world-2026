## NumPy - ndarray
### The ndarray (N-dimensional array) is the core data structure in NumPy.
 It supports multiple dimensions, making it ideal for scientific computing, data analysis, and machine learning.

### 🔹 Key Attributes of ndarray

I explored the most important properties used to describe array structure:

- **shape** → dimensions of the array
- **ndim** → number of dimensions
- **size** → total number of elements
- **dtype** → data type of elements
- **itemsize** → memory size per element

### Example:

```python
arr_2D = np.array([[1, 2, 3], [4, 5, 6]])  # 2D array


print('Looking at arr_2D key attributes:\n')
print("Shape:", arr_2D.shape)
print("Dimensions:", arr_2D.ndim)
print("Size:", arr_2D.size)
print("Data type:", arr_2D.dtype)
print("Item size:", arr_2D.itemsize)
print("\n")
# arr_3D 
arr_3D = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])  # 3D array
print('Looking at arr_3D key attributes:\n')
print("Shape:", arr_3D.shape)
print("Dimensions:", arr_3D.ndim)   
print("Size:", arr_3D.size)
print("Data type:", arr_3D.dtype)
print("Item size:", arr_3D.itemsize)

```

### Explanation of NumPy Array Attributes

####  `arr_2D`

- **Shape `(2, 3)`**  
  The array has **2 rows and 3 columns**, meaning it is a 2‑dimensional structure similar to a small table.

- **Dimensions (`ndim = 2`)**  
  The array has **2 axes** (rows and columns), so it is a **2D array**.

- **Size (`size = 6`)**  
  The total number of elements in the array.  
  Calculated as:  
  `2 × 3 = 6`.

- **Data type (`dtype = int64`)**  
  All elements are stored as **64‑bit integers**.

- **Item size (`itemsize = 8`)**  
  Each element occupies **8 bytes** of memory  
  (`int64 = 8 bytes`).

---

####  `arr_3D`

- **Shape `(2, 2, 2)`**  
  The array has **3 dimensions**, which can be thought of as **2 blocks**, each containing a **2 × 2 matrix**.

- **Dimensions (`ndim = 3`)**  
  The array has **3 axes**, making it a **3D array**.

- **Size (`size = 8`)**  
  The total number of elements in the array.  
  Calculated as:  
  `2 × 2 × 2 = 8`.

- **Data type (`dtype = int64`)**  
  All elements are stored as **64‑bit integers**, same as the 2D array.

- **Item size (`itemsize = 8`)**  
  Each element still occupies **8 bytes** of memory.

---

###  Key Takeaway

- **`shape`** describes the structure of the array  
- **`ndim`** tells how many dimensions (axes) the array has  
- **`size`** gives the total number of elements

##  Indexing and Slicing

###  Basic Indexing

I learned how to access individual elements in a NumPy array using index positions


```python

arr = np.array([10, 20, 30, 40, 50])
print(arr[2])  # 30
```

### Slicing
Slicing allows extracting a sub-array using a range of indices.

Slicing allows extracting a **portion of an array** by specifying a `start` and `end `range.

```python
# arr = np.array([10, 20, 30, 40, 50])
print(arr[1:4])  # Start: 1, Stop: 4,  by default Step: 1 [20 30 40]
```

## From this example, I learned that:

- Slicing starts at **index** `1` **inclusive**
- Stops at **index** `4` **exclusive**
- Uses the **default** step of `1`

This is useful for extracting a continuous range of values from the array.

```python
# arr = np.array([10, 20, 30, 40, 50])
print(arr[::3])  # Start: 0, Stop: end, Step: 3 [10 40]
```

### In this case:

- The *slice starts* from the **beginning** of the array
- It selects every **third element**

This helped me understand how slicing can be used to downsample data or select values at regular intervals.

```python
# arr = np.array([10, 20, 30, 40, 50])
print(arr[1:4:2])  # Start: 1, Stop: 4, Step: 2 [20 40]
```

## Here, I learned that:

- The slice starts at *index* `1`
- Stops before *index* `4`
- Takes *every* `second` element

This form of slicing is helpful when I want selective extraction within a specific range.

Indexing starts from 0, so arr[2] returns the third element.
